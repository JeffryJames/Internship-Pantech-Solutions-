```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/titanicsurvival.csv")

print(data)
data.head(10)
data.shape


# Mapping Text ( CONVERTING MALE TO 1 AND FEMALE TO 0 )

income_set = set(data['Sex'])
data['Sex'] = data['Sex'].map({'female':0, 'male':1}).astype(int)

print(data)


X = data.iloc[:,:-1]
print(X)

y = data.iloc[:,-1]
print(y)

X.columns[X.isna().any()]

X.Age = X.Age.fillna(X.Age.mean())

print(X['Age'])
X.columns[X.isna().any()]
 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.25, random_state = 0)

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)


pclassNo = int(input("Enter pclassNo : "))
gender = int(input("Enter Gender : "))
age = int(input("Enter Age : "))
fare = float(input("Enter Fare : "))
person = [[pclassNo,gender,age,fare]]
result = model.predict(person)
print(result)

if result == 1:
    print("The Person Might Have Survived")
else:
    print("The Person Might Have Not Survived")

y_pred = model.predict(X_test)
print(np.column_stack((y_pred,y_test)))

from sklearn.metrics import accuracy_score
Acc = (accuracy_score(y_test, y_pred)*100)
print("ACCURACY : ",Acc)

 
```
